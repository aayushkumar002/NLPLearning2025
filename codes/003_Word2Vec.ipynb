```python
model.most_similar("King")
#most_similar() is used to find words that are similar to the given word based on the word embeddings learned by the model.
# In the output, vector values represent the similarity scores between the input word and the similar words found by the model.
#how it works:
# The model calculates the cosine similarity between the vector representation of the input word and the vectors of all other words in the vocabulary.
#How Cosine Similarity Works:
# Cosine similarity measures the cosine of the angle between two non-zero vectors in a multi-dimensional space.
# It is calculated using the formula:   cosine_similarity(A, B) = (A · B) / (||A|| * ||B||)
# where A and B are the vectors, "·" denotes the dot product, and ||A|| and ||B|| are the magnitudes (lengths) of the vectors.
#For example, if you input the word "king", the model will return a list of words that are semantically similar to "king", along with their similarity scores.
# Load the pre-trained Word2Vec model (Google News vectors)
# Find words similar to "king"
```
