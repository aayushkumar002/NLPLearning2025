```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
path = r'D:\pyproj\venv\codes\NLPLearning2025\files'

for file in os.listdir(path):
    file_path = os.path.join(path, file)

    if os.path.isfile(file_path) and file.endswith(".txt"):
        with open(file_path, 'r', encoding='cp1252') as f:
            data = f.read()
            for sentence in sent_tokenize(data):
                story.append(simple_preprocess(sentence))

print(f"Total sentences: {len(story)}")
```

```python
story[0:5]  # Display the first 5 preprocessed sentences
```

```python
## Training the Word2Vec model and "No Stop words removal" version
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(sentence):
    return [word for word in sentence if word not in stop_words]
story_nostop = [remove_stopwords(sentence) for sentence in story]
model = gensim.models.Word2Vec(
    sentences=story_nostop,
    window=10,
    min_count=2)
## above just initializes the model parameters, that means, window there means how many words to consider left and right of the target word and min_count is the minimum frequency of words to consider
model.build_vocab(story_nostop)
## here we are building the vocabulary from the preprocessed sentences, i.e., making a dictionary of words with their frequencies
print(model)
```

```python
#time to train the model
model.train(story, total_examples=model.corpus_count, epochs=30)
```

```python
model.wv.most_similar('daenerys')  # Example to find most similar words to 'daenerys'
```

```python
model.wv.doesnt_match("jon arya tyrion cersei".split())  # Example to find the word that doesn't match
##why not jon? because jon and arya are stark siblings, tyrion is lannister but close to starks, cersei is lannister and enemy of starks  
```

```python
model.wv.similarity('cersei', 'tyrion')  # Example to find similarity between 'cersei' and 'tyrion'
```

```python
model.wv.similarity('arya', 'sansa')  # Example to find similarity between 'sansa' and 'arya'
```

```python
model.wv.get_normed_vectors() # Get normalized word vectors
## Now i want to know the word behind the 1st vector
index_to_word = model.wv.index_to_key
index_to_word  # Get the word corresponding to the first vector
## FYI Model is trained on non - stop words version of the story
```

```python
# Reduce dimensions to 3D using PCA. That means we are converting the word vectors to 3 dimensions so that we can visualize them
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Reduce dimensions to 3D using PCA
word_vectors = model.wv.get_normed_vectors()
pca = PCA(n_components=3)
reduced_vectors = pca.fit_transform(word_vectors)
print(reduced_vectors.shape)  # Should be (vocab_size, 3)
# pca is used here to reduce the dimensions of the word vectors to 3D for visualization purposes
```

```python
import plotly.express as px
### i want to label the points with the words and color code them too
words = index_to_word[200:300]
fig = px.scatter_3d(reduced_vectors[200:300], x=0, y=1,z=2, text=words,color=words)
fig.update_traces(textposition='top center')
fig.show()
## So first we preprocess the text data by tokenizing sentences and words, then we train a Word2Vec model on the preprocessed data. After training, we can use the model to find similar words, check for odd one out, and measure similarity between words. Finally, we visualize the word vectors in 3D space using PCA for dimensionality reduction.
## Now in plotly we first took words as variable and the passed index_to_word[200:300] to get first 100 words from the vocabulary to label and color code the points in the scatter plot.
## then in px.scatter_3d we passed reduced_vectors[200:300] to get the first 100 reduced vectors corresponding to the first 100 words. x=0, y=1, z=2 means we are using the first, second and third dimensions of the reduced vectors for the x, y and z axes of the 3D scatter plot. text=words means we are labeling each point with the corresponding word from the words variable. color=words means we are color coding the points based on the words variable, so each unique word gets a different color in the plot.
## in pca.fit_transform(word_vectors) we are transforming the original word vectors to 3 dimensions using PCA. This reduced dimensionality allows us to visualize the relationships between words in a 3D scatter plot.
## fig.show() displays the interactive 3D scatter plot with words labeled and color coded.
```
