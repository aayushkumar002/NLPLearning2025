```python
##########Bag of Words (BoW) Feature Extraction##########
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'text': ['people love coding',
             'love is coding',
             'i love coding',
             'coding is fun'], 'output':[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
bow = cv.fit_transform(df['text'])
print(cv.vocabulary_)
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
print(bow.toarray())
```

```python
##############n-grams Feature Extraction##############
```

```python
df2 = pd.DataFrame({
    'text': ['people love coding',
                'love is coding',
                'i love coding',
                'coding is fun'], 'output':[1,1,0,0]})

from sklearn.feature_extraction.text import CountVectorizer
cv2 = CountVectorizer(ngram_range=(1,1)) # unigrams
bow2 = cv2.fit_transform(df2['text'])
print(bow2.toarray())
print("Unigrams Vocabulary: ", cv2.vocabulary_)

cv3 = CountVectorizer(ngram_range=(2,2)) # bigrams
bow3 = cv3.fit_transform(df2['text'])
print(bow3.toarray())
print("Bigrams Vocabulary: ", cv3.vocabulary_)
```

```python
###############TF-IDF Feature Extraction###############
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
# tfidf
# print(tfidf.idf_)
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
